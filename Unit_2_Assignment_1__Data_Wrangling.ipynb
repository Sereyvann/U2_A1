```python
import sklearn 
import numpy as np
import pandas as pd
```

```python
#1.Load the dataset

dataset = pd.read_csv("https://raw.githubusercontent.com/Sereyvann/netflix_title.csv-/main/netflix_titles.csv")
dataset.head(5)
```

```python
#2.Dimensions
dataset.shape
```

```python
dataset.describe()
```

```python
#3.Check the variable(Numerical/Catagorical)
numerical_var = dataset.columns[dataset.dtypes != 'object']
dataset[numerical_var].head(5)
```

```python
categorical_var = dataset.columns[dataset.dtypes == 'object']
dataset[categorical_var].head(5)
```

```python
#4.Missing values
dataset.isnull().head(5)
```

```python
#5.Find out the percentage of missing values
dataset.isnull().sum()/len(dataset)
```

```python
#6.Replace
dataset.fillna('NA')
```

```python
#7.Fillup the missing values
dataset['director'].replace('NA', 'Special')
```

```python
#8.Create the rows and columns
row_insert = pd.Series(['s168', "Bromance", "BB", "SV", "NongNong", "China", "23-May-22", 2022,	"5 stars",	"120mn",	"Movies",	"Two best friend in high school"], index = ["show_id",	"type",	"title",	"director",	"cast",	"country",	"date_added",	"release_year",	"rating",	"duration",	"listed_in",	"description"])
dataset.append(row_insert, ignore_index=True)
```

```python
#9.Choose the exact rows from the dataset
dataset.iloc[:5]
```

```python
#10.Rename the columns
dataset.rename(columns={'show_id' : 'movie_id'})
```

```python
#11.Find the min, max, sum, count
print('Min: ', dataset['release_year'].min())
print('Max: ', dataset['release_year'].max())
print('Count: ', dataset['release_year'].count())
print('Sum: ', dataset['release_year'].sum())
```

```python
#12.Delete the rows and columns
#Columns
dataset.drop('date_added', axis=1)
```

```python
#Row
dataset.drop([0], axis=0)
```
